### Merge to one dataframe

```python
vni['Ngày'].min(), vni['Ngày'].max()
```

```python
vni.head(5)
```

```python
btc_usd.head(5)
```

```python
gold.head(5)
```

```python
btc_usd = btc_usd.merge(vni[['Ngày', 'Lần cuối']], on='Ngày', how='left', suffixes=('', '_vni'))
```

```python
btc_usd.head(5)
```

```python
btc_usd = btc_usd.merge(gold[['standard_time', 'price_mean']], left_on='Ngày', right_on='standard_time', how='left', suffixes=('', '_gold'))
```

```python
btc_usd.head(5)
```

```python
# rename columns for clarity
btc_usd.rename(columns={'Lần cuối': 'BTC_USD', 'Lần cuối_vni': 'VN_Index', 'Lần cuối_USD_VND': 'USD_VND', 'price_mean': 'Gold_Price'}, inplace=True)
```

```python
# drop redundant columns
btc_usd.drop(columns=['Mở', 'Cao', 'Thấp', 'KL', '% Thay đổi'], inplace=True)
```
